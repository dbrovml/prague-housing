## Environment

```python
#
from IPython.display import display, HTML, Image
from pandas_profiling import ProfileReport
import pandas as pd
import numpy as np
```

## Load inputs

```python
df = pd.read_parquet("./.data/features")
print(df.columns)
print(df.shape)
```

## Make pandas profile report

```python
%%capture output
path = "./report.html"
profile = ProfileReport(
    df, samples=None,
    interactions=None,
    missing_diagrams=None,
)

profile.to_file(output_file=path)
```
